# PYTHON PROJECT : PASSWORD GENERATOR

```python
import random
import string
import tkinter as tk
from tkinter import messagebox

def generate_password():
    try:
        min_length = int(length_entry.get())  # Get the minimum length
        numbers = num_var.get()  # Get checkbox value for numbers
        special_characters = special_var.get()  # Get checkbox value for special characters
        
        letters = string.ascii_letters
        digits = string.digits
        special = string.punctuation

        characters = letters
        if numbers:
            characters += digits
        if special_characters:
            characters += special

        pwd = ""
        meets_criteria = False
        has_number = False
        has_special = False

        while not meets_criteria or len(pwd) < min_length:
            new_char = random.choice(characters)
            pwd += new_char

            if new_char in digits:
                has_number = True
            elif new_char in special:
                has_special = True

            meets_criteria = True
            if numbers:
                meets_criteria = has_number
            if special_characters:
                meets_criteria = meets_criteria and has_special
        
        password_entry.delete(0, tk.END)
        password_entry.insert(0, pwd)  # Display generated password
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter a valid number for length.")

# Creating the GUI window
root = tk.Tk()
root.title("Password Generator")
root.geometry("400x300")

# Labels and Inputs
tk.Label(root, text="Enter Minimum Length:").pack(pady=5)
length_entry = tk.Entry(root)
length_entry.pack(pady=5)

num_var = tk.BooleanVar()
tk.Checkbutton(root, text="Include Numbers", variable=num_var).pack(pady=5)

special_var = tk.BooleanVar()
tk.Checkbutton(root, text="Include Special Characters", variable=special_var).pack(pady=5)

# Generate Button
tk.Button(root, text="Generate Password", command=generate_password).pack(pady=10)

# Output Field
tk.Label(root, text="Generated Password:").pack(pady=5)
password_entry = tk.Entry(root, width=30)
password_entry.pack(pady=5)

# Run the GUI
root.mainloop()
```
